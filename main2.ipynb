```python
import os
import pandas as pd  # for storing text and embeddings data
from openai import OpenAI  # for calling the OpenAI API
from openai.types.chat import ChatCompletionMessageParam

# models
EMBEDDING_MODEL = "text-embedding-3-large"
GPT_MODEL = "gpt-3.5-turbo"
```

```python
client = OpenAI(api_key=os.environ["OPENAI_API_KEY"])
```

```python
def get_example_qa():
    """
    Get user question and assistant answer for examples from files.
    """
    qa = []
    with open("data/example_short_question.txt", "r") as f:
        qa.append(f.read())
    with open("data/example_short_answer.txt", "r") as f:
        qa.append(f.read())
    return qa
```

```python
def ask(question: str, model: str = GPT_MODEL):
    q, a = get_example_qa()
    message = f"""
    Example question: {q}
    Example answer: {a}

    The above is an example question and answer. Given the intent below, generate the code.
    {question}
    """

    messages: list[ChatCompletionMessageParam] = [
        {
            "role": "system",
            "content": "You will be provided with intent, and your task is to generate the code.",
        },
        {"role": "user", "content": message},
    ]
    response = client.chat.completions.create(
        model=model, messages=messages, temperature=0.7
    )
    return response.choices[0].message.content
```

```python
ask("What is the capital of France?")
```

```python
print(ask("Transfer USDC to '0x123' on Polygon."))
```

```python
print(ask("Send 0.1 USDC to '0x123' and send 0.2 to '0x456' on Base"))
```

```python
print(ask("Split 0.1 USDC to half and then send to '0x123' & '0x456' on Base"))
```
